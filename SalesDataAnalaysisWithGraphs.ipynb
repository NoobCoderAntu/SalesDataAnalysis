```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Showing Data Table:


```python
data = pd.read_csv('/content/company_sales_data.csv')
data
```

#Question 1 : Read Total profit of all months and show it using a line plot.


```python
plt.xlabel('Month number')
plt.ylabel('Total Profit in dollar')
plt.xticks(data['month_number'])
plt.plot(data['month_number'],data['total_profit'])
```

#Question 2 : Get total profit of all months and show line plot with the following Style
properties
Generated line plot must include following Style properties: –
(i) Line Style dotted and Line-color should be red
(ii) Show legend at the lower right location.
(iii) X label name = Month Number
(iv) Y label name = Sold units number
(v) Add a circle marker.
(vi) Line marker color as read
(vii) Line width should be 3

```python
plt.plot(data['month_number'],data['total_profit'], label = 'Company profit data', color = 'r', marker = 'o', markerfacecolor = 'b', linestyle = 'dotted', linewidth = 3 )
plt.xlabel('Month number')
plt.ylabel('Sold unit number')
plt.legend(loc='lower right')
```

#Question 3 : Read all product sales data and show it using a multiline plot

```python
plt.plot(data['month_number'], data['facecream'],   label = 'FACE CREAM Sales Data', marker='o',linestyle = '--', linewidth=2)
plt.plot(data['month_number'], data['facewash'],   label = 'FACE WASH Sales Data',  marker='o', linestyle = '--',linewidth=2)
plt.plot(data['month_number'], data['toothpaste'], label = 'TOOTH PASTE Sales Data', marker='o',linestyle = '--', linewidth=2)
plt.plot(data['month_number'], data['bathingsoap'], label = 'BATHING SOAP Sales Data', marker='o',linestyle = '--', linewidth=2)
plt.plot(data['month_number'], data['shampoo'], label = 'SHAMPOO Sales Data', marker='o',linestyle = '--', linewidth=2)
plt.plot(data['month_number'], data['moisturizer'], label = 'MOISTURIZER Sales Data', marker='o',linestyle = '--', linewidth=2)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title('Sales data')
plt.xticks(data['month_number'])
plt.yticks([2000, 4000, 6000, 8000, 10000, 12000, 14000, 16000, 18000, 20000, 22000])
```

#Question 4 : Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(data['month_number'],data['toothpaste'], label = 'Sales Data of Tooth Paste', color = 'r' )
plt.legend( loc = 'upper left')
plt.xlabel('Month Number')
plt.ylabel('Number of unit Tooth Paste Sold')
plt.title('Tooth paste Sales data')
plt.xticks(data['month_number'])
```

#Question 5 : Read face cream and face wash product sales data and show it using the bar chart

```python
plt.bar([ w-0.2 for w in data['month_number']], data['facecream'] , width = 0.2, label = 'FACE CREAM SALES DATA', align = 'edge', color = 'g')
plt.bar([ w+0.2 for w in data['month_number']], data['facewash'] , width = -0.2, label = 'FACE WASH SALES DATA', align = 'edge', color = 'y')

plt.title('Facewash and facecream sales data')
plt.xticks(data['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.yticks([0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000])
```

#Question 7 : Read the total profit of each month and show it using the histogram to see the most common profit ranges.

```python
profit = [ 150000, 175000, 200000, 225000, 250000, 275000, 300000, 325000, 350000, 375000, 400000, 425000]
plt.hist(data['total_profit'], profit, label = 'TOTAL PROFIT DATA', color = 'g')

plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper right')
plt.xticks(data['total_profit'])
plt.title('TOTAL PROFIT DATA')
```

#Question 8 : Calculate total sale data for last year for each product and show it using a Pie chart.

```python
products = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesItems   = [data['facecream'].sum(), data['facewash'].sum(), data['toothpaste'].sum(), data['bathingsoap'].sum(), data['shampoo'].sum(), data['moisturizer'].sum()]

plt.axis("equal")
plt.pie(salesItems, labels = products, autopct='%1.1f%%')
plt.title('SALES DATA OF ALL PRODUCTS')
```

#Question 9: Read Bathing Soap and  Facewash of all months and display it using the Subplot.

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
fig.suptitle('SALES DATA OF BATHING SOAP AND FACE WASH', fontsize=20)

ax[0].plot(data['month_number'], data['bathingsoap'], color = 'y', marker = 'o')
ax[0].set_title('Sales Data of Bathing Soap', fontsize = 15)

ax[1].plot(data['month_number'], data['facewash'], color = 'b', marker = 'o')
ax[1].set_title('Sales Data of Facewash', fontsize = 15)

plt.xticks(data['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
```

#Question 10 : Read all product sales data and show it using the stack plot.

```python
plt.plot([],[], label = 'FACE CREAM')
plt.plot([],[], label = 'FACE WASH')
plt.plot([],[], label = 'TOOTH PASTE')
plt.plot([],[], label = 'BATHING SOAP')
plt.plot([],[], label = 'SHAMPOO')
plt.plot([],[], label = 'MOISTURIZER')

plt.stackplot( data['month_number'], data['facecream'], data['facewash'], data['toothpaste'], data['bathingsoap'], data['shampoo'], data['moisturizer'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.xticks(data['month_number'])
plt.yticks([0, 5000, 10000, 15000, 20000,25000, 30000,35000,40000, 45000, 50000])
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
```
